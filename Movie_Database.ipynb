```python
%%sql
SELECT 
COUNT(DISTINCT GMAP.movie_id) as Thriller_Love
FROM genremap as GMAP
JOIN keywordmap as kMAP
ON GMAP.movie_id = KMAP.movie_id
JOIN keywords as KWORD
ON KWORD.keyword_id = KMAP.keyword_id
JOIN genres as GEN
ON GEN.genre_id = GMAP.genre_id
WHERE genre_name = 'Thriller' AND keyword_name LIKE '%love%';
```

```python
%%sql
SELECT 
count (DISTINCT characters) as Character_Played
FROM casts as CA
JOIN actors as AC
ON CA.actor_id = AC.actor_id
WHERE AC.actor_name = 'Vin Diesel';
```

```python
%%sql
SELECT genre_name
FROM movies as MV
JOIN genremap as GMAP
ON MV.movie_id = GMAP.movie_id
JOIN genres as GEN
ON GEN.genre_id = GMAP.genre_id
WHERE MV.title = "The Royal Tenenbaums";
```

```python
%%sql
SELECT 
     production_company_name,
     AVG(MV.popularity) as AVG_p
FROM productioncompanymap as PCM
JOIN movies as MV
ON PCM.movie_id = MV.movie_id
JOIN productioncompanies as PC
ON PC.production_company_id = PCM.production_company_id
GROUP BY production_company_name
ORDER BY AVG_p DESC 
LIMIT 3;
```

```python
%%sql
SELECT 
     genre_name,
     AVG(MV.popularity) as AVG_p
FROM genremap as GMAP
JOIN movies as MV
ON GMAP.movie_id = MV.movie_id
JOIN genres as GEN
ON GMAP.genre_id = GEN.genre_id
GROUP BY genre_name
ORDER BY AVG_p
LIMIT 1;
```

```python
%%sql
SELECT award, COUNT(*) AS num_nominations
FROM Oscars
WHERE award LIKE '%Actor%' 
   OR award LIKE '%Actress%'
   OR award LIKE 'Special Achievement Award'
   OR award LIKE 'Best Picture'
GROUP BY award
ORDER BY num_nominations DESC
LIMIT 1;
```

```python
%%sql
DROP VIEW IF EXISTS Alan_Rickman_Movies;
CREATE VIEW Alan_Rickman_Movies AS 
SELECT title, 
release_date, 
tagline, 
overview 
FROM Movies 
LEFT JOIN Casts 
ON Casts.movie_id = Movies.movie_id 
Left JOIN Actors 
ON Casts.actor_id = Actors.actor_id 
WHERE Actors.actor_name = 'Alan Rickman';
```

```python
%%sql
SELECT * 
FROM Alan_Rickman_Movies
LIMIT 10; 
```

```python
%%sql
SELECT
    strftime('%Y', release_date) AS release_year,
    COUNT(*) AS movie_count
FROM Movies
WHERE release_date IS NOT NULL
GROUP BY release_year
ORDER BY movie_count DESC
LIMIT 5;
```

```python
%%sql
SELECT
    strftime('%Y', M.release_date) AS release_year,
    M.title,
    M.release_date
FROM Movies AS M
WHERE strftime('%Y', M.release_date) IN (
      SELECT strftime('%Y', release_date)
      FROM Movies
      WHERE release_date IS NOT NULL
      GROUP BY strftime('%Y', release_date)
      ORDER BY COUNT(*) DESC
      LIMIT 5
)
ORDER BY release_year, M.title;
```
